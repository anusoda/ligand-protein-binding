```python
import pandas as pd
import numpy as np
from fragmentembedding import FragmentEmbedder
from rdkit import Chem
```

```python
def is_valid_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return mol is not None
```

```python
bdb = pd.read_csv("../BindingDB/BindingDB_minimal_cols_human_proteins_1uM_or_less_EC50_IC50_Kd_Ki_subclass_annotated.csv")
```

```python
bdb_smiles = bdb['Ligand SMILES'].tolist()
invalid = []
for i, val in enumerate(bdb_smiles):
    if not is_valid_smiles(val):
        invalid.append(i)
invalid
```

```python
block = 1000
invalid_blocks = set()
for inv in invalid:
    invalid_blocks.add(inv//block)
invalid_blocks
```

```python
def iterate_block(fe, index):
    X = []
    for i in range(index*block, min((index+1)*block, len(bdb_smiles))):
        if i in invalid:
            continue
        try:
            Y = fe.transform([bdb_smiles[i]])
            X.append(Y)
        except ValueError:
            invalid.append(i)
    return np.concatenate(X)
```

```python
fe = FragmentEmbedder()
X = []
for i in range(0, (len(bdb_smiles)//block)+1):
    print(block*i)
    if i in invalid_blocks:
        X.append(iterate_block(fe, i))
    else:
        try:
            X.append(fe.transform(bdb_smiles[block*i:block*(i+1)]))
        except ValueError:
            X.append(iterate_block(fe, i))
final = np.concatenate(X)
np.save("bdb_embeddings.npy", final)
final.shape
```
